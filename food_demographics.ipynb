```python
groupedByKnowledge = df.groupby('Food_Knowledge')
kls = []

for knowledge_group in groupedByKnowledge:
    kls.append(knowledge_group[0])

counts = []
    
knowledge_cuisine_scores = dict.fromkeys(kls)

knowledgeScores = []
for knowledge_group in groupedByKnowledge:   
    count = 0
    countryScores = []
    for country in countries:
        
        scoreSum = 0
        for entry in knowledge_group[1][country]:

            count = count + 1
            scoreSum = scoreSum + entry

        countryScores.append(scoreSum)

    counts.append(count)
    countryScoresDict = dict(zip(countries, countryScores))
    knowledgeScores.append(countryScoresDict)
overallScores = dict(zip(kls, knowledgeScores))

# number of entries in each group
kcounts = dict(zip(kls, counts))
# pprint.pprint (overallScores)

aggregate_weight = len(countries)*5

for x in overallScores:
    for y in overallScores[x]:
        overallScores[x][y] = overallScores[x][y]/kcounts[x]

pprint.pprint(overallScores)
```
