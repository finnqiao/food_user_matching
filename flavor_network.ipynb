```python
import re 
pattern = re.compile(r'[^A-Za-z\s]') 

def word_weights(str_):
    str_ = pattern.sub('',str_) 
    alist = str_.lower().split(" ")
    length = len(alist)
    default=0
    
    dict={}
    for i in alist:
        if dict.get(i, default)==default:
            dict[i]=1/length
        else:
            dict[i]+=1/length

    return dict
```
