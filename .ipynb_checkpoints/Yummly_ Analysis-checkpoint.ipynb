```python
cuisines = []
flavors = []
ingredients = []
name = []
for filename in os.listdir('metadata27638'):
    if filename.endswith('.json'):
        data = json.load(open('metadata27638/'+filename))
        cuisines.append(data['attributes']['cuisine'][0])
        flavors.append(data['flavors'])
        ingredientStr = ''
        for ingred in data['ingredientLines']:
            ingredientStr = ingredientStr + ingred
        ingredients.append(ingredientStr)
        name.append(data['name'])
        
df = pd.DataFrame({'cuisine':cuisines,'name':name,'flavors':flavors,'ingredients':ingredients})
```

```python
df.count()
```

```python
yummly = df
yummly = yummly.drop(yummly[(yummly.cuisine == 'Kid-Friendly')].index)
```

```python
yummly_grouped = yummly.groupby('cuisine')
for x in yummly_grouped:
    print (x)
```

```python
yummly_words = dict.fromkeys(['african', 'american', 'chinese', 'french', 'indian', 'italian', 'japanese', 'korean', 'latin_american', 'mediterranean', 'mexican', 'other_asian', 'other_european', 'spanish', 'thai', 'vietnamese', 'middle_eastern'])
for key in yummly_words:
    yummly_words[key] = ''

flavors = ['Sweet','Sour','Salty','Piquant','Bitter','Meaty']


cuisine_flavors = []
yummly_cuisines = []
    
for cuisine in yummly_grouped:
    count = 0
    flavor_dict = dict.fromkeys(flavors)
    for key in flavor_dict:
        flavor_dict[key] = 0
        
    for flavorObj in cuisine[1].flavors:
        if flavorObj == {}:
            flavorObj = dict.fromkeys(flavors)
            for key in flavor_dict:
                flavorObj[key] = 0
        elif len(flavorObj) != 6:
            continue
        else:
            count = count + 1
            for key in flavor_dict:
                flavor_dict[key] = flavor_dict[key] + flavorObj[key]
    for key in flavor_dict:
        if flavor_dict[key] != 0:
            flavor_dict[key] = flavor_dict[key]/count
    yummly_cuisines.append(cuisine[0])
    cuisine_flavors.append(flavor_dict)
flavor_map = dict(zip(yummly_cuisines, cuisine_flavors))
flavor_map['korean'] = flavor_map['other_asian']
flavor_map['middle_eastern'] = flavor_map['african']
pprint(flavor_map)
    
#     for x in cuisine[1].ingredients:
#         for y in x:
#             kaggle_words[cuisine[0]]=kaggle_words[cuisine[0]] + y + ' '
#     print('done')
```
