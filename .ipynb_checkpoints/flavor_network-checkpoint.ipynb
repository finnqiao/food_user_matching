```python
cuisine_list = []
ingred_list = []
with open('flavor_network.csv', newline='') as csvfile:
    r = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for line in r:
        cuisine = line[0].split(',')[0]
        cuisine_list.append(cuisine)
        
        ingred = (' ').join(line[0].split(',')[1:]).replace('_',' ')
        ingred_list.append(ingred)

        
df = pd.DataFrame({'cuisines':cuisine_list,'ingredients':ingred_list})
```

```python
# print(x.ingredients['African'])
ingred_word_cloud = []
for i in range(len(cuisine_list)):
    ingred_word_cloud.append(x.ingredients[i])

print(len(cuisine_list))
print(len(ingred_word_cloud))
ingred_df = pd.DataFrame({'cuisines':cuisine_list,'ingreds':ingred_word_cloud})
ingred_df.head()
```

```python
word_scores = dict.fromkeys(ingred_df.cuisines)
for i in range(len(ingred_df.cuisines)):
    tracker = ''
    words = ''
    for x in ingred_df.ingreds[i]:
        if tracker == ' ' and x == ' ':
            continue
        words = words + x
        tracker = x
    ingred_df.ingreds[i] = words
    word_scores[ingred_df.cuisines[i]] = word_weights(ingred_df.ingreds[i])

pp.pprint (word_scores)
```

```python
new_word_scores = {}
for key, item in word_scores.items():
    if '' in item:
        del item['']
    new_word_scores[key] = sorted(item.items(), key=lambda x: x[1], reverse = True)
pp.pprint(new_word_scores)
```

```python
import re 
pattern = re.compile(r'[^A-Za-z\s]') 

def word_weights(str_):
    str_ = pattern.sub('',str_) 
    alist = str_.lower().split(" ")
    length = len(alist)
    default=0
    
    dict={}
    for i in alist:
        if dict.get(i, default)==default:
            dict[i]=1/length
        else:
            dict[i]+=1/length

    return dict
```
