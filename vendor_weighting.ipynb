```python
def cross_weight(d1, d2):
    default=1 
    ret=0;
    for i in d1.keys():
        if(d2.get(i,default)==default):
            continue
        else:
            ret+=d1[i]*d2[i]

    return ret
```

```python
import re 
pattern = re.compile(r'[^A-Za-z\s]') 


def word_weights(str_):
    str_ = pattern.sub('',str_) 
    alist = str_.lower().split(" ")
    length = len(alist)
    default=0
    
    dict={}
    for i in alist:
        if dict.get(i, default)==default:
            dict[i]=1/length
        else:
            dict[i]+=1/length

    return dict
```

```python
def simplify(str_):
    str_ = pattern.sub('',str_) 
    alist = str_.lower().split(" ")
    str_ = '_'.join(alist)
    return str_
```
