```python
def get_info(str_):
#     from roaming hunger get name, about, attributes, menu
    rhlink = rhLinks[str_]
    req = Request(rhlink, headers={'User-Agent': 'Mozilla/5.0'})
    r = urlopen(req).read()
    soup = BeautifulSoup(r, "html.parser")
#     gets name
    name = soup.select('.title span')[0].text
#     gets about
    aboutStr = ''
    for tag in soup.select('.about p'):
        aboutStr = aboutStr + tag.text
    about = aboutStr
#     gets attributes
    savory = len(soup.select('.tags .col-sm-3 .savory')) 
    sweet = len(soup.select('.tags .col-sm-3 .sweet'))
    vegetarian = len(soup.select('.tags .col-sm-3 .vegetarian'))
    attrs = dict({'savory':savory, 'sweet':sweet, 'vegetarian':vegetarian})
#     gets menu
    menuStr = ''
    for x in soup.select('.menu h3'):
        menuStr = menuStr + ' ' + x.text
    for x in soup.select('.menu p'):
        menuStr = menuStr + ' ' + x.text
    
    infoObj = dict.fromkeys(information)

    eventVars = ['location', 'specLocation', 'times']
    timeVars = ['day', 'date', 'time']

    timeObj = dict.fromkeys(timeVars)
    timeObj['day'] = []
    timeObj['date'] = []
    timeObj['time'] = []
    eventObj = dict.fromkeys(eventVars)
    eventObj['times'] = timeObj
    infoObj['schedule'] = eventObj
    
    infoObj['name'] = name
    infoObj['about'] = about
    infoObj['attribute'] = attrs
    infoObj['menu'] = menuStr

#     from offthegrid get menu and schedule
    from selenium import webdriver 
    browser = webdriver.Chrome(executable_path="/Users/finn/Munchmates/food_user_matching/chromedriver") #replace with .Firefox(), or with the browser of your choice
    url = otgLinks[str_]

    browser.get(url) #navigate to page behind login
    innerHTML = browser.execute_script("return document.body.innerHTML")
    browser.quit()
    soup = BeautifulSoup(innerHTML, "html.parser")
    schedule = []
    
#     gets a list of scheduled events
    for location in soup.select('.events-wrap .grid-item-card'):
        timeObj = dict.fromkeys(timeVars)
        timeObj['day'] = []
        timeObj['date'] = []
        timeObj['time'] = []
        eventObj = dict.fromkeys(eventVars)
        eventObj['times'] = timeObj
        print(eventObj)
        eventObj['location']= location.select('.location')[0].text
        eventObj['specLocation'] = location.select('.event-where')[0].text
        for time in location.select('ul li'):
            if len(time.select('.upcoming__day'))>0:
                eventObj['times']['day'].append(time.select('.upcoming__day')[0].text)
            if len(time.select('.upcoming__day'))>0:
                eventObj['times']['date'].append(time.select('.upcoming__date')[0].text)
            if len(time.select('.upcoming__time'))>0:
                eventObj['times']['time'].append(time.select('.upcoming__time')[0].text)
        schedule.append(eventObj)
    
#     gets another copy of menu
    menuStr2 = ''
    for x in soup.select('.menu .grid-item-card header'):
        menuStr2 = menuStr2 + x.text + ' '
    for x in soup.select('.menu .grid-item-card .content'):
        menuStr2 = menuStr2 + x.text + ' '
        
    infoObj['schedule'] = schedule
    infoObj['menu'] = infoObj['menu'] + menuStr2
    
#     no info for trucks yet for chef info, star count, review count, cuisine, price, location
    infoObj['chef-info'] = ''
    infoObj['stars-count'] = 0
    infoObj['review-count'] = 0
    infoObj['cuisine'] = ''
    infoObj['price'] = 1
    infoObj['location'] = None
    infoObj['type'] = 'truck'
    return infoObj

print(otgLinks['el-sur'])
get_info('el-sur')
```

```python
trucksInfo = {}
from concurrent.futures import ThreadPoolExecutor, as_completed

with ThreadPoolExecutor(max_workers=10) as executor: 
    future_to_truck = {executor.submit(get_info, truck): truck for truck in keys}
    for future in as_completed(future_to_truck):
        truck = future_to_truck[future]
        trucksInfo[truck] = future.result()
```
