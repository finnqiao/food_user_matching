```python
cuisines = []
flavors = []
ingredients = []
for filename in os.listdir('metadata27638'):
    if filename.endswith('.json'):
        data = json.load(open('metadata27638/'+filename))
        cuisines.append(data['attributes']['cuisine'][0])
        flavors.append(data['flavors'])
        ingredientStr = ''
        for ingred in data['ingredientLines']:
            ingredientStr = ingredientStr + ingred
        ingredients.append(ingredientStr)
        
df = pd.DataFrame({'cuisine':cuisines,'flavors':flavors,'ingredients':ingredients})
df.head()
```
